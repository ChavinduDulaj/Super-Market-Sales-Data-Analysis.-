```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# List of columns you want to label encode
columns_to_encode = ['Sale_or_Return', 'Discount_(Yes/No)']


# Encode and replace each column in the DataFrame
for col in columns_to_encode:
    df[col] = label_encoder.fit_transform(df[col])

# Now, the original columns have been replaced with label encoded values

```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Time_seconds'] = df['Time'].dt.hour * 3600 + df['Time'].dt.minute * 60 + df['Time'].dt.second + df['Time'].dt.microsecond / 1e6
df = df.drop(['Date','Time'], axis = 'columns')
```

```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Check if GPU is available
if tf.test.gpu_device_name():
    print('GPU found')
else:
    print("No GPU found")

# Load sales data
sales_data = df

# Reduce data size
reduced_size = int(len(sales_data) * 0.2)  # Reduce to 20%
sampled_data = sales_data.sample(n=reduced_size, random_state=1)

# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(sampled_data.values.reshape(-1, 1))

# Split data into train and test sets
train_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - train_size
train_data, test_data = scaled_data[0:train_size], scaled_data[train_size:]

# Prepare data for LSTM model
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Reshape input to be [samples, time steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Build LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
with tf.device('/GPU:0'):  # Specify GPU device
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=64, verbose=1)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Plotting
plt.plot(sampled_data.index[time_step:len(train_predict)+time_step], train_predict, label='Train Predictions')
plt.plot(sampled_data.index[len(train_predict)+(2*time_step)+1:len(train_predict)+len(test_predict)+(2*time_step)+1], test_predict, label='Test Predictions')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Prediction using LSTM')
plt.legend()
plt.show()
```
