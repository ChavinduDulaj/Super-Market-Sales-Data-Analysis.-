Importing libraries fro data cleaning and preprocessing

```python
import pandas as pd
import numpy as np
```

```python
item_cat_df = pd.read_csv('dataset/annex1.csv')
sales_df = pd.read_csv('dataset/annex2.csv')
wholesale_df = pd.read_csv('dataset/annex3.csv')
avg_loss_df = pd.read_csv('dataset/annex4.csv')
```

```python
item_cat_df
```

```python
item_cat_df.columns = item_cat_df.columns.str.replace(" ", "_")
```

```python
sales_df.columns = sales_df.columns.str.replace(" ", "_")
```

```python
sales_df.isnull().sum()
```

```python
for column in sales_df.columns:
    unique_values = sales_df[column].unique()
    print(f"Unique values in column '{column}':")
    print(unique_values)
    print()
```

```python
wholesale_df.columns = wholesale_df.columns.str.replace(" ", "_")
```

```python
wholesale_df.isnull().sum()
```

```python
for column in wholesale_df.columns:
    unique_values = wholesale_df[column].unique()
    print(f"Unique values in column '{column}':")
    print(unique_values)
    print()
```

```python
avg_loss_df.columns = avg_loss_df.columns.str.replace(" ", "_")
```

Data Merging:

```python
print(sales_df['Date'].min(), sales_df['Date'].max())
print(wholesale_df['Date'].min(), wholesale_df['Date'].max())
```

```python
sales_wholesale_df = pd.merge(sales_df, wholesale_df, how = 'left', on = ['Item_Code', "Date"])
sales_wholesale_df.head()
```

```python
sales_wholesale_df.isnull().sum()
```

```python
sales_wholesale_cat_df = pd.merge(sales_wholesale_df, item_cat_df, how = 'left', on = ['Item_Code'])
sales_wholesale_cat_df.head()
```

```python
sales_wholesale_cat_df.isnull().sum()
```

```python
data = pd.merge(sales_wholesale_cat_df, avg_loss_df, how = 'left', on = ['Item_Code','Item_Name'])
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.to_csv('dataset\merged_data.csv', index=False)
```
